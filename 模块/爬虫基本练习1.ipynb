```python
import re
import requests

url = 'http://blog.csdn.net'
headers = {'User-Agent':'Mozilla/5.0 (X11; Linux x86_64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/55.0.2883.87 Safari/537.36'}
data = requests.get(url, headers=headers).text
pat = '<h3  class="tracking-ad" data-mod="popu_254"><a href="(.*?)"'
link = re.findall(pat, data)
for i in range(0, len(link)):
    thislink = link[i]
    r = requests.get(thislink, headers=headers)
    with open(r'/home/deepin/blogcsdn/'+ str(i+1) + '.html', 'wb') as f:
        f.write(r.content)


    print ('正在爬取第'+ str(i+1),'页', thislink)
```
